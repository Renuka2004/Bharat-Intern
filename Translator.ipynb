```python
from translate import Translator
```

```python

def read_languages_from_file(filename, encoding='utf-8'):
    with open(filename, 'r', encoding=encoding) as file:
        languages = [line.strip() for line in file]
    return languages

def translate_text(input_text, source_language, target_language):
    translator = Translator(from_lang=source_language, to_lang=target_language)
    translation = translator.translate(input_text)
    return translation

# Read languages from the file
languages = read_languages_from_file('languages.txt')

# Display available languages
print("Available languages:")
for index, language in enumerate(languages, 1):
    print(f"{index}. {language}")

source_language_name = input("Enter the source language name: ").strip().capitalize()
target_language_name = input("Enter the target language name: ").strip().capitalize()

if source_language_name not in languages or target_language_name not in languages:
    print("Invalid language name.")
else:
    source_language_code = source_language_name.lower()
    target_language_code = target_language_name.lower()

    input_text = input("Enter the text to translate: ")
    translated_text = translate_text(input_text, source_language_code, target_language_code)
    print(f"Original text: {input_text}")
    print(f"Translated text: {translated_text}")
```
